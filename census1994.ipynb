# 2: Aggregation & Filtering & Rank
In this task, we will perform some very high level aggregation and filtering operations. Then, we will apply ranking on the results for some tasks. Pandas has a convenient and powerful syntax for aggregation, filtering, and ranking. DO NOT write a for loop. Pandas has built-in functions for all tasks. 

```python
#Task 2-a: Find out the sum of Captial Gain for people with education level as Bachelors and HS-Grad.
sum_capital_gain_bachelors = df_census[df_census["education"] == " Bachelors"]["capital-gain"].sum()
sum_capital_gain_HS_Grad = df_census.groupby('education')['capital-gain'].sum()[' HS-grad']
print ("Task 2-a: The sum of capital gain for education level as bachelors is %s and as HS-Grad is %s" 
       % (sum_capital_gain_bachelors, sum_capital_gain_HS_Grad))


#Task 2-b: Find out the total number of people surveyed in months may, october and december.
#Create a new column for 'Survey_Month' by using 'Date' column
#write the code for extracting the month from the date column here

###############begin your code here

df_census["Survey_Month"] = df_census['Date'].apply(lambda x:(datetime.datetime.strptime(x,'%m/%d/%Y').date().strftime("%B")))

###############send you code here

num_surveys_may = df_census.loc[df_census.Survey_Month == 'May' , 'Survey_Month'].count()
num_surveys_october = df_census.loc[df_census.Survey_Month == 'October' , 'Survey_Month'].count()
num_surveys_december = df_census.loc[df_census.Survey_Month == 'December' , 'Survey_Month'].count()
print ("\n\nTask 2-b: The total number of surveys in may is %s, in october is %s, and in december is %s" 
       % (num_surveys_may, num_surveys_october, num_surveys_december))


#Task 2-c: Let us now use multiple filtering criteria
# Find out the total number of surveys in september and november with workclass as private and age less than 50.
num_surveys_september = df_census[(df_census["Survey_Month"] == "September") & (df_census["WorkClass"] == " Private") & (df_census["Age"] < 50)].count()[0]
num_surveys_november = df_census[(df_census["Survey_Month"] == "November") & (df_census["WorkClass"] == " Private") & (df_census["Age"] < 50)].count()[0]
print ("\n\nTask 2-c: The total number of surveys that meet the given conditions in september is %s and in november is %s" 
       % (num_surveys_september, num_surveys_november))


#Task 2-d: Find out 3 least surveyed education categories, print their names and corresponding number of surveys for periods January-June and July-December.
months = ["January","February","March","April","May","June"]
months2 = ["July","August","September","October","November","December"]
top3_least_surveyed_Jan_June = {}
temp_df = df_census[df_census["Survey_Month"].isin(months)].groupby("education").count().sort_values(by=["Survey_Month"]).iloc[0:3,:]
for x in temp_df.index:
    top3_least_surveyed_Jan_June[x] = temp_df.loc[x][0]
top3_least_surveyed_July_December = {}
temp_df = df_census[df_census["Survey_Month"].isin(months2)].groupby("education").count().sort_values(by=["Survey_Month"]).iloc[0:3,:]
for x in temp_df.index:
    top3_least_surveyed_July_December[x] = temp_df.loc[x][0]
print ("\n\nTask 2-d: \nThe top 3 least surveyed education categories in January-June: \n%s \n\nThe top 3 least surveyed education categories in July-December: \n%s" 
                     % (top3_least_surveyed_Jan_June,top3_least_surveyed_July_December))



#Task 2-e: Find out top 5 native-countries besides United-States, print their names and number of surveys belonging to each.
top5_most_surveyed_native_countries = {}
native = [" United-States"]
temp_df = df_census[~(df_census["native-country"].isin(native))].dropna().groupby("native-country").count().sort_values(by="Date").iloc[-5:,0:1]
for x in temp_df.index:
    top5_most_surveyed_native_countries[x] = temp_df.loc[x][0]
print ("\n\nTask 2-e: \nThe top 5 most surveyed native countries : \n%s" 
                     % (top5_most_surveyed_native_countries))


#Task 2-f: Find out Top-5 native-countries with the most number of samples belonging to class >50K
top5_native_countries = {}
temp_df = df_census[df_census["class"] == " >50K"].groupby("native-country").count().sort_values(by="Date").iloc[-5:,0:1]
for x in temp_df.index:
    top5_native_countries[x] = temp_df.loc[x][0] 
print ("\n\nTask 2-f: \nThe top 5 native countries with the most number of surveys with class >50K: \n%s" 
                     % (top5_native_countries))
```

# 3: Visualization
In this task, we will perform a number of visualization tasks to get some intuition about the data. Visualization is a key component of exploration. You can choose to use either Matplotlib or Seaborn for plotting. The default figures generated from Matplotlib might look a bit ugly. So you might want to try Seaborn to get better figures. Seaborn has a variety of styles. Feel free to experiment with them and choose the one you like. We have earmarked 10 points for the aesthetics of your visualizations.

```python
#setting the figure size
plt.rcParams["figure.figsize"] = [7, 6]
#Task 3-a: Draw a histogram for total number of surveys taken each month. Dislpay months with their corresponding numbers(Eg: January is 1) 
#########################begin code for Task 3-a
df_census["Survey_Month"] = df_census['Date'].apply(lambda x:(datetime.datetime.strptime(x,'%m/%d/%Y').date().strftime("%m")))
graph1 = df_census.groupby("Survey_Month").count()["Date"]
months = graph1.index
plt.xlabel('Survey Months',fontsize=9,color='black')
plt.ylabel('Total Number of Surveys', fontsize=9,color='black')
plt.title("Surveys taken each month",color='black')
plt.style.use('seaborn-ticks')
plt.bar(list(graph1.index),list(graph1.values),color='rebeccapurple',edgecolor='cyan')
plt.show()
#########################end code for Task 3-a


#Task 3-b: Draw a vertical bar chart for total number of surveys taken for each gender for each month. Display months with their corresponding names.
# Remember to make the bar chart into a vertical bar chart
#########################begin code for Task 3-b
df_census["Survey_Month"] = df_census['Date'].apply(lambda x:(datetime.datetime.strptime(x,'%m/%d/%Y').date().strftime("%B")))
order = ['January', 'February', 'March','April','May','June','July', 'August', 'September','October','November','December']
color = ['salmon', 'maroon']
graph2 = df_census.groupby('gender').Survey_Month.value_counts().unstack(0).loc[order].plot(kind='bar',color=color, legend=True, fontsize=12)
plt.xlabel("Survey Months",fontsize=9,color='black')
plt.ylabel('Total Number of Surveys', fontsize=9,color='black')
plt.title("Surveys taken each month for each gender",color='black')
plt.xticks(rotation=45)
plt.show()
#########################end code for Task 3-b

#Task 3-c: Draw a horizontal bar chart for number of surveys taken with respect to age feature keeping the age interval as 15.
# Remember to make the bar chart into a horizontal bar chart
#########################begin code for Task 3-c
bins = np.arange(df_census["Age"].min(), df_census["Age"].max(), 15)
graph3 = df_census.groupby([ pd.cut(df_census["Age"],bins)]).count()["Date"]
list_index = []
for x in graph3.index:
    list_index.append("{}-{}".format(x.left,x.right))
plt.xlabel('Total Number of Surveys', fontsize=9)
plt.ylabel('Age Interval', fontsize=9)
plt.title("Surveys take with respect to age")
plt.barh(list_index,list(graph3.values), facecolor=(0.1843, 0.3098, 0.3098))
plt.show()
#########################end code for Task 3-c


#Task 3-d: Draw a "vertical" bar chart that lists the top-5 native-countries based on the number of samples with class >50K.
# Remember to make the bar chart into a vertical bar chart
#########################begin code for Task 3-d
graph4 = df_census[df_census["class"] == " >50K"].groupby("native-country").count().sort_values(by="Date").iloc[-5:,:]["Date"]
plt.xlabel('Top-5 native-countries', fontsize=9)
plt.ylabel('Total Number of Surveys', fontsize=9)
plt.title("Top 5 countires with class > 50k")
plt.xticks(rotation=45)
plt.bar(list(graph4.index),list(graph4.values))
plt.show()
#########################end code for Task 3-d


#Task 3-e: Now repeat Task 3-d based on education (again top-5)
#########################begin code for Task 3-e
graph5 = df_census[df_census["class"] == " >50K"].groupby("education").count().sort_values(by="Date").iloc[-5:,:]["Date"]
plt.xlabel('Top-5 education category', fontsize=9)
plt.ylabel('Total Number of Surveys', fontsize=9)
plt.title("Surveys take with respect to education")
plt.xticks(rotation=45)
plt.bar(list(graph5.index),list(graph5.values),color='mediumseagreen')
plt.style.use('seaborn-dark-palette')
plt.show()
#########################end code for Task 3-e


#Task 3-f: Draw a scatter plot for age vs hours per week.
#########################begin code for Task 3-f
graph6 = df_census[["Age","hours-per-week"]].groupby("Age").count()
data_dict = graph6.to_dict()["hours-per-week"]
area = np.pi*4

# Plot
plt.scatter(data_dict.keys(), data_dict.values(), s=area, alpha=0.5, color='black')
plt.title('Age vs Hours/Week')
plt.xlabel('Age', fontsize=9)
plt.ylabel('Hours/Week', fontsize=9)
plt.show()
#########################end code for Task 3-f


#Task 3-g: Draw a line chart showing average capital gain for each education category.
# X-axis : education category, Y-axis : the avg capital gain
#########################begin code for Task 3-g
graph7 = df_census.groupby("education").mean()["capital-gain"]
plt.xlabel('Education Category', fontsize=9)
plt.ylabel('Average captital gain', fontsize=9)
plt.title("Education vs Avg Capital gain")
plt.xticks(rotation=45)
plt.plot(list(graph7.index),list(graph7.values),color='crimson')
plt.show()
#########################end code for Task 3-g


#Task 3-h: Draw a 'horizontal' bar chart for the top-5 most common occupation. 
#########################begin code for Task 3-h
#using seaborn
graph10 = df_census.groupby("occupation").count().sort_values(by = "Date").iloc[-5:,:]
graph10.reset_index(level=0, inplace=True)
sns.set_color_codes("pastel")
sns.barplot(x="Age", y="occupation", data=graph10,
            label="Count", color="b")

# Add a legend and informative axis label
plt.legend(ncol=1, loc="lower right", frameon=True)
plt.xlabel("Count", fontsize=9)
plt.ylabel("Occupation", fontsize=9)
plt.yticks(rotation=45)
plt.title("Top-5 Common Occupation")
plt.show()
#########################end code for Task 3-h


#Task 3-i: Draw a 'horizontal' bar chart for the top-5 most common workclass. 
#########################begin code for Task 3-i
graph9 = df_census.groupby("WorkClass").count().sort_values(by = "Date").iloc[-5:,:]["Date"]
plt.xlabel('Count', fontsize=9,color='black')
plt.ylabel('Working Class', fontsize=9,color='black')
plt.title("Top-5 Working Class",color='black')
plt.yticks(rotation=45)
plt.style.use('seaborn-dark-palette')
plt.barh(list(graph9.index),list(graph9.values),color='midnightblue')
plt.show()
#########################end code for Task 3-i
```

Hence we are going to use World Maps get to get more detailed infromation of data.

```python
# import the library
#!pip install folium
#create a file from https://github.com/python-visualization/folium/blob/master/examples/data/world-countries.json
import folium
temp_df = df_census
temp_df["native-country"] = temp_df["native-country"].str.replace(" ","")
#either download the json file from the URL or import from the world-countries.json file provided along with the assignment
import os
country_geo =  os.path.join('world-countries.json')
```

```python
##This graph explains the average life expectency based on the regions 
graph1 = temp_df.groupby("native-country").mean()
graph1.reset_index(level=0, inplace=True)
m = folium.Map(location=[0, 0], zoom_start=2)
m.choropleth(
 geo_data=country_geo,
 name='choropleth',
 data=graph1,
 columns=['native-country', 'Age'],
 key_on='feature.properties.name',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Average Life Expectency '
)
folium.LayerControl().add_to(m)
m
```

```python
##Going in detail into the average life expectancy, we will be plotting for male average life expectency, followed by female average life expectency.
graph2 = temp_df[temp_df["gender"] == " Male"].groupby(["native-country"]).mean()
graph2.reset_index(level=0, inplace=True)

m = folium.Map(location=[0, 0], zoom_start=2)
m.choropleth(
 geo_data=country_geo,
 name='choropleth',
 data=graph2,
 columns=['native-country',"Age"],
 key_on='feature.properties.name',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Male Average Life Expectency '
)
folium.LayerControl().add_to(m)
m
```

```python
graph3 = temp_df[temp_df["gender"] == " Female"].groupby(["native-country"]).mean()
graph3.reset_index(level=0, inplace=True)

m = folium.Map(location=[0, 0], zoom_start=2)
m.choropleth(
 geo_data=country_geo,
 name='choropleth',
 data=graph3,
 columns=['native-country',"Age"],
 key_on='feature.properties.name',
 fill_color='YlGn',
 fill_opacity=0.7,
 line_opacity=0.2,
 legend_name='Female Average Life Expentency '
)
folium.LayerControl().add_to(m)
m
```
